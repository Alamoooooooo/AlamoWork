# CV Score

```python
from torch.utils.data import Dataset, DataLoader
class CustomDataset(Dataset):
    def __init__(self, df,   accelerator):
        self.features = torch.FloatTensor(df[CONFIG.feature_cols].values).to(accelerator)
        self.labels = torch.FloatTensor(df["responder_6"].values).to(accelerator)
        self.weights = torch.FloatTensor(df["weight"].values).to(accelerator)
    
    def __len__(self):
        return len(self.labels)
    
    def __getitem__(self, idx):
        x = self.features[idx]
        y = self.labels[idx]
        w = self.weights[idx]
        return x, y, w

valid_dataset = CustomDataset(valid, "cpu")
valid_dataloader = DataLoader(valid_dataset, batch_size=8192, shuffle=False, num_workers=6)
```

```python
y_valid = []
y_pred_valid = []
w_valid = []

with torch.no_grad():  # 不计算梯度，节省内存和计算资源
    for batch in tqdm(valid_dataloader):
        inputs, labels, w = batch
        y_pred_valid_temp = np.zeros(labels.shape)
        for model in models:
            model.eval()
  # 获取输入、标签和权重（如果需要）
            
            # 将数据转移到与模型相同的设备
            inputs = inputs.to("cuda")
            
            # 执行模型预测
            _, outputs = model(inputs)

            # 还要成一个系数
            y_pred_valid_temp += outputs.cpu().numpy()
            # 记录预测结果和真实标签
        y_pred_valid.append(y_pred_valid_temp)  # 转回CPU，方便后续处理
        y_valid.append(labels.numpy())
        w_valid.append(w.numpy())
            

# 将所有预测结果和标签拼接在一起
y_pred_valid = np.concatenate(y_pred_valid, axis=0)
y_valid = np.concatenate(y_valid, axis=0)
w_valid = np.concatenate(w_valid, axis=0)
```

```python
valid_score = r2_score(y_valid, y_pred_valid, sample_weight=w_valid )
valid_score
```

```python
# 画图，对每个valid中存在的symbol_id，画出实际值和预测值的比较，并打印出rmse和mae
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error

def plot_predictions(y_valid, y_pred, symbol_id):
    plt.figure(figsize=(10, 6))
    plt.plot(np.cumsum(y_valid ), label=f'Actual Values for Symbol ID: {symbol_id}', linewidth=1)
    plt.plot(np.cumsum(y_pred ), label=f'Predicted Values for Symbol ID: {symbol_id}' , linewidth=1)
    plt.xlabel('Data Points')
    plt.ylabel('Values')
    plt.title(f'Actual vs. Predicted Values for Symbol ID: {symbol_id}')
    plt.legend()
    plt.show()

def calculate_metrics(y_valid, y_pred):
    rmse = np.sqrt(mean_squared_error(y_valid, y_pred))
    mae = mean_absolute_error(y_valid, y_pred)
    return rmse, mae


```
